```python
import pandas as pd

# Load submissions
heavy = pd.read_csv("submission_heavy.csv")
balanced = pd.read_csv("submission_balanced.csv")

# Copy heavy model as base
ensemble = heavy.copy()

# Blend severity
ensemble["severity"] = (
    0.7 * heavy["severity"] +
    0.3 * balanced["severity"]
).round().astype(int)

# Save ensemble
ensemble.to_csv("submission_ensemble.csv", index=False)

print("Ensemble submission created")
ensemble.head()
```
